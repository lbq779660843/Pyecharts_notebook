# Ⅰ.柱状图(Bar)

## 基础示例
第一段代码包含了在柱形图绘制过程中基本上需要使用的API，并且做出了简单的注释;代码还未排版
其它类型的图像参数与此共用

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType    #使用主题
from pyecharts.commons.utils import JsCode


x = ["衬衫", "毛衣", "领带", "裤子", "风衣", "高跟鞋", "袜子"]
y1 = [114, 55, 27, 101, 125, 27, 105]
y2 = [57, 134, 137, 129, 145, 60, 49]
y3 = [99, 196, 56, 29, 115, 90, 149]
# V1 版本开始支持链式调用
bar = (
    Bar(    #初始化主题
            #{"theme": ThemeType.MACARONS}
                init_opts=opts.InitOpts(
                    #开启动画,参考https://blog.csdn.net/W_H_M_2018/article/details/105565792
                    animation_opts=opts.AnimationOpts(
                    animation_delay=1000, animation_easing="elasticOut")
                    #设置背景
                    #bg_color={"type": "pattern", "image": JsCode("img"), "repeat": "no-repeat"}
            
        )
        
       )
    
    #******此处是添加数据，注意很多例子在'.add_'格式的函数前两项写有'series_name='和'yaxis_data='，记得去掉!******
    .add_xaxis(x)
    .add_yaxis("商家A", y1, gap="0%")#gap为不同 系列柱 间的距离
    .add_yaxis("商家B", y2, category_gap="80%", is_selected=False)
    .add_yaxis("商家C", y3, stack="stack1", category_gap="60%")
    #.reversal_axis() #对调xy轴
    #"category_gap=80%"表示条目之间间隔占比为80%(最后一个生效), is_selected设置设否默认显示可以靠点击恢复, stack表示堆叠
    
    
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False),#设置序列选项
                     markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(coord=[x[2], y1[2]], value=y1[2]), #自定义指示☞点
                                                             opts.MarkPointItem(type_="max", name="最大值"), #设置标点,此处是最大值
                                                             opts.MarkPointItem(type_="min", name="最小值"),
                                                             opts.MarkPointItem(type_="average", name="平均值"),
                                                                               ]),
                     markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(y=50, name="yAxis=50"),#自定义指示☞线
                                                          opts.MarkLineItem(type_="min", name="最小值"),
                                                          opts.MarkLineItem(type_="max", name="最大值"),
                                                          opts.MarkLineItem(type_="average", name="平均值"),]
                                                    ),
                     itemstyle_opts={#柱子->椭圆以及颜色
                        "normal": {
                            "color": JsCode(
                                """new echarts.graphic.LinearGradient(0, 0, 0, 1, [{
                        offset: 0,
                        color: 'rgba(0, 244, 255, 1)'
                    }, {
                        offset: 1,
                        color: 'rgba(0, 77, 167, 1)'
                    }], false)"""
                        ),
                        "barBorderRadius": [30, 30, 30, 30],
                        "shadowColor": "rgb(0, 160, 221)",
                                }
                                    }
                    )
    
    
    .set_global_opts(#用链式调用形式配置
        title_opts=opts.TitleOpts(title="某商场销售情况",title_textstyle_opts=opts.TextStyleOpts(color="white"),),#设置标题及文字风格
        yaxis_opts=opts.AxisOpts(name="Y轴表示价格"),
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15,formatter="{value} /月")),         #设置x轴翻转及格式
    )
    .set_global_opts( #用字典形式配置
        title_opts={"text": "另一种配置方式", "subtext": "次级标题"},   
        datazoom_opts={opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")},                              #数据整体放大
        #datazoom_opts={opts.DataZoomOpts(type_="inside")},                                                  #数据仅内部放大
        #datazoom_opts={opts.DataZoomOpts()},                                                                #数据仅水平放大
        #datazoom_opts={opts.DataZoomOpts(orient="vertical")},                                               #数据仅纵向放大
        toolbox_opts={opts.ToolboxOpts()},                                                                   #加载工具箱
        #brush_opts={opts.BrushOpts()},                                                                       #添加右侧滑动条
        
        )
)
# 保存为html文件
bar.render('test_1.html')


#官方写法
from pyecharts.faker import Faker    #使用内置虚假数据
bar_official = (
    Bar({"theme": ThemeType.MACARONS})#初始化主题
    .add_xaxis(Faker.choose())#Faker.choose()随机调用一类标签
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    .set_global_opts(title_opts=opts.TitleOpts(title="某商场销售情况", subtitle="2020年"))#标题配置项
    .render("bar_base.html")#这行建议尽量写在外部以防报错
)
```
