# Ⅴ.日历

## 基础示例

```python
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar


begin = datetime.date(2020, 1, 1)
end = datetime.date(2020, 12, 31)
#print(datetime.timedelta(days=2))
data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]#timedelta 理解为量纲为'天'的整形数
    for i in range((end - begin).days + 1)
]
#print(data)

c = (
    Calendar()
    .add("", data, calendar_opts=opts.CalendarOpts(range_="2020"))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Calendar-2020年微信步数情况"),
        visualmap_opts=opts.VisualMapOpts(
            max_=20000,
            min_=500,
            orient="horizontal",
            is_piecewise=True,
            pos_top="230px",
            pos_left="100px",
        ),
    )
    #.render("calendar_base.html")
)
c.load_javascript()
c.render_notebook()
```

## 日历(热力图形式)

```python
begin = datetime.date(2020, 1, 1)
end = datetime.date(2020, 12, 31)
data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
    for i in range((end - begin).days + 1)
]

cld_htm = (
    Calendar()
    .add(
        series_name="",
        yaxis_data=data,
        calendar_opts=opts.CalendarOpts(
            pos_top="120",
            pos_left="30",
            pos_right="30",#定位
            range_="2020",#确定年份
            yearlabel_opts=opts.CalendarYearLabelOpts(is_show=False),
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(pos_top="30", pos_left="center", title="2020年步数情况"),
        visualmap_opts=opts.VisualMapOpts(
            max_=20000, min_=500, orient="horizontal", is_piecewise=False#最大最小值，方向, 分段图还是热力图
        ),
    )
    #.render("calendar_heatmap.html")
)
cld_htm.load_javascript()
cld_htm.render_notebook() 
```

## 标签设置

```python
begin = datetime.date(2017, 1, 1)
end = datetime.date(2017, 12, 31)
data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
    for i in range((end - begin).days + 1)
]

bqsz = (
    Calendar()
    .add(
        "",
        data,
        calendar_opts=opts.CalendarOpts(
            range_="2017",
            #日标签和月标签都设置为中文格式
            daylabel_opts=opts.CalendarDayLabelOpts(name_map="cn"),
            monthlabel_opts=opts.CalendarMonthLabelOpts(name_map="cn"),
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Calendar-2017年微信步数情况(中文 Label)"),
        visualmap_opts=opts.VisualMapOpts(
            max_=20000,
            min_=500,
            orient="horizontal",
            is_piecewise=True,
            pos_top="230px",
            pos_left="100px",
        ),
    )
    #.render("calendar_label_setting.html")
)
bqsz.load_javascript()
bqsz.render_notebook() 
```
